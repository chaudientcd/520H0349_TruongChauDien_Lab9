```python
#Ex1
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n_machines = 5

p_broken = 0.1

X = np.arange(0, n_machines + 1)

probabilities = binom.pmf(X, n_machines, p_broken)

# (a) Probability that 2 machines are broken in 1 session (X = 2)
prob_2_broken = probabilities[2]
print(f"The probability that 2 machines are broken in 1 session is: {prob_2_broken:.3f}")

# (b) Create a line graph representing the relationship between X and the corresponding probability
plt.plot(X, probabilities, marker='o', linestyle='-')
plt.xlabel('Number of Broken Machines (X)')
plt.ylabel('Probability')
plt.title('Probability Distribution of Broken Machines in 1 Session (Line Graph)')
plt.grid(True)
plt.show()
```

```python
#Ex2
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson


average_calls = 3


X = np.arange(1, 6)

probabilities = poisson.pmf(X, average_calls)

# (a) Probability that the center receives 2 calls in 1 minute (X = 2)

prob_2_calls = probabilities[1]
print(f"The probability that the center receives 2 calls in 1 minute is: {prob_2_calls:.3f}")

# (b) Create a line graph representing the relationship between X and the corresponding probability

plt.plot(X, probabilities, marker='o', linestyle='-')
plt.xlabel('Number of Calls (X)')
plt.ylabel('Probability')
plt.title('Probability Distribution of Calls in 1 Minute')
plt.grid(True)
plt.show()
```

```python
#Ex3
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

p_success = 0.4

X = np.arange(1, 11)

probabilities = geom.pmf(X, p_success)

# (a) Probability that a person hits the target in his third try (X = 3)
prob_3rd_try = probabilities[2]
print(f"The probability that a person hits the target in his third try is: {prob_3rd_try:.3f}")

# (b) Create a line graph representing the relationship between X and the corresponding probability
plt.plot(X, probabilities, marker='o', linestyle='-')
plt.xlabel('Number of Attempts (X)')
plt.ylabel('Probability')
plt.title('Probability Distribution of Attempts to Hit the Target')
plt.grid(True)
plt.show()
```
